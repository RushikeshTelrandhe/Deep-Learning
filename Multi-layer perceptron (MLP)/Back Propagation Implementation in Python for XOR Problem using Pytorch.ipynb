```python
# Import Libraries
import torch
import torch.nn.functional as F
```

```python
# Training Data (XOR)
X = torch.tensor([[0., 0.],[0., 1.],[1., 0.],[1., 1.]])
y = torch.tensor([[0.],[1.],[1.],[0.]])
```

```python
# Initialize Weights and Biases
torch.manual_seed(0)

input_size = 2
hidden_size = 4
output_size = 1

W1 = torch.randn(input_size, hidden_size, requires_grad=True)
b1 = torch.zeros(hidden_size, requires_grad=True)

W2 = torch.randn(hidden_size, output_size, requires_grad=True)
b2 = torch.zeros(output_size, requires_grad=True)
```

```python
# Training Loop (Feedforward + Backpropagation)
learning_rate = 0.1
epochs = 10000

for epoch in range(epochs):
    # ----- Feedforward -----
    hidden = torch.sigmoid(X @ W1 + b1)
    output = torch.sigmoid(hidden @ W2 + b2)

    # ----- Loss -----
    loss = torch.mean((y - output) ** 2)

    # ----- Backpropagation -----
    loss.backward()

    # ----- Update Weights -----
    with torch.no_grad():
        W2 -= learning_rate * W2.grad
        b2 -= learning_rate * b2.grad
        W1 -= learning_rate * W1.grad
        b1 -= learning_rate * b1.grad

        # Reset gradients
        W1.grad.zero_()
        b1.grad.zero_()
        W2.grad.zero_()
        b2.grad.zero_()

    if epoch % 4000 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item()}")
```

```python
# Testing the Network
with torch.no_grad():
    hidden = torch.sigmoid(X @ W1 + b1)
    predictions = torch.sigmoid(hidden @ W2 + b2)

print("Predictions after training:")
print(predictions)
```
